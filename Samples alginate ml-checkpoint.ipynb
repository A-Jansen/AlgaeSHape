## Importing data and creation of dataframes

```python
file_alginate = 'alginate ml.csv'
df_alg= pd.read_csv(file_alginate)
df_alg.head()
```

```python
df_alg.info()
```

DC % = 100 ( B – A) / A

DC = Dimensional Change

A = Original Dimension

B = Dimension after Laundering

Shrinkage is denoted as ‘-’ which is decrease in dimensions

Elongation is denoted as ‘+’ which is increase in dimensions.
HMS. (2014, January 28). DIMENSIONAL STABILITY TO WASHING AND DRYING TEST FOR FABRICS AND GARMENTS «. Retrieved 3 June 2020, from http://garmentstech.com/dimensional-stability-to-washing-and-drying-test-for-fabrics-and-garments/

```python
df_alg['Shrinkage height']= ((df_alg['Height after']- df_alg['Height'])/df_alg['Height'] *100)
df_alg['Shrinkage width']= ((df_alg['Width after']-df_alg['Width'])/df_alg['Width'] *100)
df_alg.head(2)
```

## Visualisation

```python
fig =df_alg.plot(x='Alginate (ml)', y=['Shrinkage height', 'Shrinkage width']).set_title('Shrinkage dependent on volume gel/ 10 cm')
fig
```

```python
fig.figure.savefig('shrinkage-ml.png', dpi=800)
```

## Statistics

```python
X_reg= df_alg[['Alginate (ml)']].copy()
y_reg= df_alg[['Shrinkage width']].copy()
reg = LinearRegression()
reg.fit(X_reg, y_reg);
coef = reg.coef_
intercept = reg.intercept_
coef, intercept
```

```python
X2 = sm.add_constant(X_reg)
est = sm.OLS(y_reg, X2)
est2 = est.fit()
print(est2.summary())
```

```python
ax= sns.regplot(X_reg['Alginate (ml)'], y_reg['Shrinkage width'])
ax.set(xlabel='Alginate (ml)/ 10 cm')
ax.set_title('Regression plot shrinkage width');
```

```python
ax.figure.savefig('reg plot ml alginate width.png', dpi=800)
```
