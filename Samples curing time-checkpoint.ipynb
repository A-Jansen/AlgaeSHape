## Importing data and creating dataframes

```python
file_alg= 'alginate curing time.csv'
df_alg= pd.read_csv(file_alg)
df_alg.head()
```

Calculating the stretch and shrink perecentage for the samples with thick alginate

DC % = 100 ( B – A) / A


DC = Dimensional Change

A = Original Dimension

B = Dimension after Laundering


Shrinkage is denoted as ‘-’ which is decrease in dimensions


Elongation is denoted as ‘+’ which is increase in dimensions.

HMS. (2014, January 28). DIMENSIONAL STABILITY TO WASHING AND DRYING TEST FOR FABRICS AND GARMENTS «. Retrieved 3 June 2020, from http://garmentstech.com/dimensional-stability-to-washing-and-drying-test-for-fabrics-and-garments/


```python
df_alg['Shrinkage height']= ((df_alg['Height after']- df_alg['Height'])/df_alg['Height'] *100)
df_alg['Shrinkage width']= ((df_alg['Width after']-df_alg['Width'])/df_alg['Width'] *100)
df_alg.head(2)
```

```python
fig =df_alg.plot(x='Curing time (min)', y=['Shrinkage height', 'Shrinkage width']).set_title('Curing time');
fig
```

```python
fig.figure.savefig('curing time.png', dpi=800)
```

## Statistics


```python
X_reg= df_alg[['Curing time (min)']].copy()
y_reg= df_alg[['Shrinkage width']].copy()
reg = LinearRegression()
reg.fit(X_reg, y_reg);
coef = reg.coef_
intercept = reg.intercept_
coef, intercept
```

```python
X2 = sm.add_constant(X_reg)
est = sm.OLS(y_reg, X2)
est2 = est.fit()
print(est2.summary())
```

```python
fig2= sns.regplot(X_reg['Curing time (min)'], y_reg['Shrinkage width']).set_title('Regression plot curing time width');
```

```python
fig2.figure.savefig('reg plot curing time width.png', dpi=800)
```
