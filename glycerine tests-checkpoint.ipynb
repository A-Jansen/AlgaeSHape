Calculating the stretch and shrink perecentage for the samples with thick alginate

DC % = 100 ( B – A) / A


DC = Dimensional Change

A = Original Dimension

B = Dimension after Laundering


Shrinkage is denoted as ‘-’ which is decrease in dimensions


Elongation is denoted as ‘+’ which is increase in dimensions.

HMS. (2014, January 28). DIMENSIONAL STABILITY TO WASHING AND DRYING TEST FOR FABRICS AND GARMENTS «. Retrieved 3 June 2020, from http://garmentstech.com/dimensional-stability-to-washing-and-drying-test-for-fabrics-and-garments/

```python
df_alg['Shrinkage height']= ((df_alg['Height after']- df_alg['Height'])/df_alg['Height'] *100)
df_alg['Shrinkage width']= ((df_alg['Width after']-df_alg['Width'])/df_alg['Width'] *100)
df_alg.head(2)
```

```python
df_pergl['Shrinkage height']= ((df_pergl['Height after']- df_pergl['Height before'])/df_pergl['Height before'] *100)
df_pergl['Shrinkage width']= ((df_pergl['Width after']-df_pergl['Width before'])/df_pergl['Width before'] *100)
df_pergl.head(2)
```

## Statistics

T-test to compare if the percentages results are different

```python
d_5h = sm.stats.DescrStatsW(df_pergl[df_pergl['% Glycerine']== 5]['Shrinkage height'])
d_10h = sm.stats.DescrStatsW(df_pergl[df_pergl['% Glycerine']== 10]['Shrinkage height'])
d_20h = sm.stats.DescrStatsW(df_pergl[df_pergl['% Glycerine']== 20]['Shrinkage height'])
d_30h = sm.stats.DescrStatsW(df_pergl[df_pergl['% Glycerine']== 30]['Shrinkage height'])
d_40h = sm.stats.DescrStatsW(df_pergl[df_pergl['% Glycerine']== 40]['Shrinkage height'])
d_50h = sm.stats.DescrStatsW(df_pergl[df_pergl['% Glycerine']== 50]['Shrinkage height'])
cm = sm.stats.CompareMeans(d_5h, d_10h)
ttest2 = cm.ttest_ind(alternative='two-sided')
ttest1 = cm.ttest_ind(alternative='larger')
ttest2[1], ttest1[1]
```

```python
cm2 = sm.stats.CompareMeans(d_10h, d_20h)
ttest2_1 = cm2.ttest_ind(alternative='two-sided')
ttest1_1 = cm2.ttest_ind(alternative='larger')
ttest2_1[1], ttest1_1[1]
```

```python
cm3 = sm.stats.CompareMeans(d_20h, d_30h)
ttest2_2 = cm3.ttest_ind(alternative='two-sided')
ttest1_2 = cm3.ttest_ind(alternative='larger')
ttest2_2[1], ttest1_2[1]
```

```python
cm4 = sm.stats.CompareMeans(d_30h, d_40h)
ttest2_3 = cm4.ttest_ind(alternative='two-sided')
ttest1_3 = cm4.ttest_ind(alternative='larger')
ttest2_3[1], ttest1_3[1]
```

```python
cm5 = sm.stats.CompareMeans(d_40h, d_50h)
ttest2_4 = cm5.ttest_ind(alternative='two-sided')
ttest1_4 = cm5.ttest_ind(alternative='larger')
ttest2_4[1], ttest1_4[1]
```

```python
cm6 = sm.stats.CompareMeans(d_5h, d_20h)
ttest2_5 = cm6.ttest_ind(alternative='two-sided')
ttest1_5 = cm6.ttest_ind(alternative='larger')
ttest2_5[1], ttest1_5[1]
```
